## Search artist name and query the URI to get top tracks

```python
artist = 'Justin Bieber'

results = sp.search(q='artist:' + artist, type='artist')
items = results['artists']['items']


items[0]['images'][0]['url']
```

```python
uri = items[0]['uri']

final = sp.artist_top_tracks(uri)

final['tracks']
```

```python
for track in final['tracks'][:10]:
    print('track: ' + track['name'])
    print('track popularity: ' + str(track['popularity']))
    print('track duration: ' + str(track['duration_ms']) + 'ms')
    print('track album: ' + track['album']['name'])
    print('release date: ' + track['album']['release_date'])
    print('audio: ' + track['external_urls']['spotify'])
    print()
```

```python
# track_names = [track["track"]["name"] for track in tracks]
track_names = [track['name'] for track in final['tracks']]
track_artists = [", ".join([artist["name"] for artist in track["artists"]]) for track in final['tracks']]
track_album = [track['album']['name'] for track in final['tracks']]
track_duration = [track['duration_ms'] for track in final['tracks']]
track_popularity = [track['popularity'] for track in final['tracks']]
track_release_date = [track['album']['release_date'] for track in final['tracks']]
track_audio = [track['external_urls']['spotify'] for track in final['tracks']]
                 
```

```python
for i in range(len(final['tracks'])):
    print(track_names[i])
    print(track_artists[i])
    print(track_album[i])
    print(track_duration[i])
    print(track_popularity[i]) 
    print(track_release_date[i])
    print(track_audio[i])
    print('')
```
